```python
folium.Choropleth(
    geo_data=geo_json_data,
    name='State Unemployment',
    data=unemployment,
    columns=['State','Unemployment'],
    key_on='feature.id',
    fill_color='BuPu',
    fill_opacity=0.7,
    line_opacity=0.5,
    legend_name='Unemployment rate (\%)',
    #threshold_scale=threshold_scale
    #reset=True
).add_to(m)

folium.LayerControl().add_to(m)
m.save('index.html')
m
```
